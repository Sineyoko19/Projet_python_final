```python
def user (x):
    """
    0 : Non consommateurs
    1 : Consommateurs
    """
    if x in ["Jamais utilisé", "Utilisé depuis une décennie"]:
        x= 0
    else :
            x= 1 
    return x
```

**Encodage des variables**

```python
from sklearn.preprocessing import LabelBinarizer
encoder=LabelBinarizer()
#Création a function that will take the dataframe and will create into the dataframe the columns Age0,Age1 (could be used for education_level too) etc

def labelBinarizing(df,case):
    encoding=encoder.fit_transform(df[case])
    encoding=pd.DataFrame(encoding)
    cols=encoding.shape[1]
    for i in range(cols):
        col_name=case+str(i)
        df[col_name]=list(encoding.loc[:,i])
        
#For education level:
        
labelBinarizing(df_ml_DP,"Education_level")
labelBinarizing(df_ml_HL,"Education_level")
labelBinarizing(df_ml_ST,"Education_level")

df_ml_DP.drop(["Education_level"],axis=1,inplace=True)
df_ml_ST.drop(["Education_level"],axis=1,inplace=True)
df_ml_HL.drop(["Education_level"],axis=1,inplace=True)
```

### Partie 3.3 : Data scaling

Les variables qui pourraient être sujettes à une normalisation dans l'étape du data preprocessing sont les variables : Neuroticism, Extraversionness, Openness,Agreability, Conscientiousness, Impulsivness et Sensation_Seeking.

**Observation 6 :** On voit bien que les valeurs des variables sont bien uniformes (Les moyennes sont de même grandeur), toutefois, afin d'avoir une échelle définit et fixe en min-max pour toutes ces colonnes, nous décidons donc faire un minmax scaler sur ces colonnes

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
# Créons une fonction qui prend un dataframe df et qui scale sur la liste de  colonne cols qui sont mis en paramètres

def Scaling(df,cols):
    for col in cols:
        if col in list(df.columns):
            df[[col]]=scaler.fit_transform(df[[col]])
            
cols=['Neuroticism', 'Extraversionness', 'Openness',
       'Agreability', 'Conscientiousness', 'Impulsiveness',
       'Sensation_Seeking']
            
Scaling(St_X_train,cols)
Scaling(Dp_X_train,cols)
Scaling(Hl_X_train,cols)
    
```

### Partie 3.4 : Modélisation

####  La regression logistique

```python
from sklearn.linear_model import LogisticRegression

score_lr=[]
model_logreg=LogisticRegression( solver='liblinear')
#On the Hallucinogen dataset
model_logreg.fit(Hl_X_train,Hl_Y_train)
score_lr.append(model_logreg.score(Hl_X_train,Hl_Y_train))

#On the depressor dataset
model_logreg.fit(Dp_X_train,Dp_Y_train)
score_lr.append(model_logreg.score(Dp_X_train,Dp_Y_train))

#On the stimulant dataset
model_logreg.fit(St_X_train,St_Y_train)
score_lr.append(model_logreg.score(St_X_train,St_Y_train))
    
```

```python
from statistics import mean
```

```python
print("Scores on simple logistic regression is : ", score_lr)
print("The mean score is",mean(score_lr))
```

 Nous allons maintenant faire une crossvalidation sur ce modèle, en variant le nombre de folds afin de pouvoir mieux évaluer la performance de la régression logistique sur nos 3 datasets.

```python
from sklearn.model_selection import cross_val_score
val_score_hl=[]
val_score_st=[]
val_score_dp=[]
val_score_gen=[]

for i in [3,5,10]:
    val_hl=cross_val_score(model_logreg,Hl_X_train,Hl_Y_train,cv=i,scoring='accuracy').mean()
    val_st=cross_val_score(model_logreg,St_X_train,St_Y_train,cv=i,scoring='accuracy').mean()
    val_dp=cross_val_score(model_logreg,Dp_X_train,Dp_Y_train,cv=i,scoring='accuracy').mean()
    val_score_hl.append(val_hl)
    val_score_st.append(val_st)
    val_score_dp.append(val_dp)
    val_score_gen.append(mean([val_hl,val_st,val_dp]))
```
