```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

from sklearn.preprocessing import LabelBinarizer

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler

import numpy as np

from sklearn.neighbors import KNeighborsClassifier


from sklearn.pipeline import Pipeline

from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer
```

```python
data=pd.read_csv("df_ml.csv")
data
```

```python
data=data.drop(["Ethnicity","Country"],axis=1)
```

```python
data["Hallucinogènes"]=data.apply(lambda x: int((x["LSD"]+ x["Mushrooms"]+ x["Ketamin"]+ x["Cannabis"]+ x["LegalH"]+ x["Ecstasy"])>0), axis = 1)
```

```python
data["Stimulants"]=data.apply(lambda x: int((x["Amphet"]+x["Coke"]+x["Ketamin"]+x["LegalH"]+x["Nicotine"]+x["Ecstasy"]+x["Crack"]+x["Semer"]+x["Vsa"])>0), axis = 1)
```

```python
data["Depresseurs"]=data.apply(lambda x: int((x["Alcohol"]+x["Heroine"]+x["Meth"]+x["Amyl"]+x["LegalH"]+x["Benzos"]+x["Ketamin"])>0), axis = 1)
```

```python
data.drop(['Alcohol',
       'Amphet', 'Amyl', 'Benzos', 'Cannabis', 'Coke', 'Crack', 'Ecstasy',
       'Heroine', 'Ketamin', 'LegalH', 'LSD', 'Meth', 'Mushrooms', 'Nicotine',
       'Semer', 'Vsa'],axis=1,inplace=True)
```

```python
df_ml_HL=data[['Age', 'Sexe', 'Education_level',
       'Neuroticism', 'Extraversionness', 'Openness', 'Agreability',
       'Conscientiousness', 'Impulsiveness', 'Sensation_Seeking',
       'Hallucinogènes']]

df_ml_ST=data[['Age', 'Sexe', 'Education_level',
       'Neuroticism', 'Extraversionness', 'Openness', 'Agreability',
       'Conscientiousness', 'Impulsiveness', 'Sensation_Seeking',
       'Stimulants']]

df_ml_DP=data[['Age', 'Sexe', 'Education_level',
       'Neuroticism', 'Extraversionness', 'Openness', 'Agreability',
       'Conscientiousness', 'Impulsiveness', 'Sensation_Seeking',
       'Depresseurs']]
```

```python
categoricals = ['Age', 'Sexe','Education_level']
binaries = ['Education_level']
numericals = ['Neuroticism', 'Extraversionness', 'Openness', 'Agreability',
       'Conscientiousness', 'Impulsiveness', 'Sensation_Seeking']
```

```python
categorical_pipe = Pipeline([('onehot', OneHotEncoder())])
```

from sklearn.base import TransformerMixin #gives fit_transform method for free
class MyLabelBinarizer(TransformerMixin):
    def __init__(self, *args, **kwargs):
        self.encoder = LabelBinarizer(*args, **kwargs)
    def fit(self, x, y=0):
        self.encoder.fit(x)
        return self
    def transform(self, x, y=0):
        return self.encoder.transform(x)


binary_pipe = Pipeline([
    ('binarize', MyLabelBinarizer())
])

```python
numerical_pipe = Pipeline([('scaler', MinMaxScaler())])
```

```python
preprocess_pipe = ColumnTransformer([
    ('cat', categorical_pipe, categoricals),
    ('num', numerical_pipe, numericals),
])
```

```python
Hl_X=df_ml_HL.drop("Hallucinogènes",axis=1)
Hl_Y=np.array(df_ml_HL[["Hallucinogènes"]]).reshape(1885,)

# For the Stimulant dataset

St_X=df_ml_ST.drop("Stimulants",axis=1)
St_Y=np.array(df_ml_ST[["Stimulants"]]).reshape(1885,)

# For the Depresseur dataset

Dp_X=df_ml_DP.drop("Depresseurs",axis=1)
Dp_Y=np.array(df_ml_DP[["Depresseurs"]]).reshape(1885,)
```

```python
import random

Hl_X_train,Hl_X_test,Hl_Y_train,Hl_Y_test=train_test_split(Hl_X,Hl_Y, test_size=0.2,random_state=100)

St_X_train,St_X_test,St_Y_train,St_Y_test=train_test_split(St_X,St_Y, test_size=0.2,random_state=100)

Dp_X_train,Dp_X_test,Dp_Y_train,Dp_Y_test=train_test_split(Dp_X,Dp_Y, test_size=0.2,random_state=100)
```

```python
knn_Hl=KNeighborsClassifier(algorithm= 'auto', n_neighbors= 85, weights= 'uniform')
knn_Dp=KNeighborsClassifier(algorithm= 'auto', n_neighbors= 55, weights= 'uniform')
knn_St=KNeighborsClassifier(algorithm= 'auto', n_neighbors= 75 , weights= 'uniform')
```

```python
model_Hl = Pipeline([
    ('pre', preprocess_pipe),
    ('knn_Hl', knn_Hl)
])

model_Dp = Pipeline([
    ('pre', preprocess_pipe),
    ('knn_Dp', knn_Dp)
])

model_St = Pipeline([
    ('pre', preprocess_pipe),
    ('knn_St', knn_St)
])
```

```python
model_Hl.fit(Hl_X_train,Hl_Y_train)

model_Dp.fit(Dp_X_train,Dp_Y_train)

model_St.fit(St_X_train,St_Y_train)
```
